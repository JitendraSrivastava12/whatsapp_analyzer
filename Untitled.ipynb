```python
# Step 1: Unzip the WhatsApp chat export
import zipfile
import os

zip_path = 'WhatsApp Chat with NITP CSE-I Unofficial Group 2023-27.zip'  # Replace with your ZIP file name
output_dir = 'unzipped_chat'

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(output_dir)

# Step 2: Load the chat text file
chat_file = None
for filename in os.listdir(output_dir):
    if filename.endswith('.txt'):
        chat_file = os.path.join(output_dir, filename)
        break

if not chat_file:
    raise FileNotFoundError("No .txt file found in ZIP archive!")

with open(chat_file, 'r', encoding='utf-8') as file:
    lines = file.readlines()
```

```python
import re
import pandas as pd

# Join the lines into one big text in case of multi-line messages
text = ''.join(lines)

# Pattern assumes 24-hour format like: 16/06/2025, 20:27 - Sender: Message
pattern = r'(\d{1,2}/\d{1,2}/\d{2,4}), (\d{1,2}:\d{2}) - (.*?): (.*?)(?=\n\d{1,2}/\d{1,2}/\d{2,4}, \d{1,2}:\d{2} - |$)'

matches = re.findall(pattern, text, re.DOTALL)

# Create DataFrame
df = pd.DataFrame(matches, columns=['Date', 'Time', 'Sender', 'Message'])

# Parse date
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Clean up whitespace
df['Message'] = df['Message'].str.strip()

# Peek at the data
print("Total messages:", len(df))
print("Participants:", df['Sender'].nunique())
```

```python
# Define keywords that typically appear in system messages
system_keywords = ['added', 'left', 'removed', 'changed', 'created', 'pinned', 'deleted', 'joined']

# Filter out messages that contain any of the system keywords
df_cleaned = df[~df['Message'].str.lower().str.contains('|'.join(system_keywords))]

# Optional: Also remove messages where sender is missing (system messages)
df_cleaned = df_cleaned[df_cleaned['Sender'].notna()]

# Reset index
df_cleaned.reset_index(drop=True, inplace=True)

# Preview cleaned data
print("Messages after removing system notifications:", len(df_cleaned))
df_cleaned.head()
```

```python
# Remove system messages where 'Sender' contains keywords or doesn't follow normal naming
system_keywords = ['added', 'removed', 'changed', 'created', 'pinned', 'left', 'deleted', 'joined']

# Keep only rows where sender doesn't contain system keywords AND doesn't start with '~' or contain 'changed their phone number'
df_cleaned = df[
    (~df['Sender'].str.lower().str.contains('|'.join(system_keywords))) &
    (~df['Sender'].str.contains('~')) &
    (~df['Sender'].str.contains('changed their phone number')) &
    (~df['Sender'].str.contains('this group'))
].copy()

df_cleaned.reset_index(drop=True, inplace=True)
```

```python
# Compute average sentiment per user
sender_sentiment = df_cleaned.groupby('Sender')['Sentiment'].mean().reset_index()

# Add sentiment label
def classify_sentiment(score):
    if score > 0.02:
        return 'Positive'
    elif score < -0.02:
        return 'Negative'
    else:
        return 'Neutral'

sender_sentiment['Sentiment_Label'] = sender_sentiment['Sentiment'].apply(classify_sentiment)
```
